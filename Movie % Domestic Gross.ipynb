```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('P4-Section6-Homework-Dataset.csv', encoding='latin1')
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
len(movies)
```

```python
movies.Genre.unique()
```

```python
# filter only requires movie Genre
movies2 = movies[movies.Genre.isin(['action', 'comedy', 'adventure','animation','drama'])]
movies2.head()
```

```python
movies.Studio.unique()
```

```python
# filter only requires movie Studio
movies3 = movies2[movies2.Studio.isin(['Buena Vista Studios', 'Sony', 'Universal','Paramount Pictures','WB','Fox'])]
movies3.head()
```

```python
print (movies3.Genre.unique())
print (movies3.Studio.unique())
print (len(movies3))
```

```python
# Define the style
plt.figure(figsize=(10,7))
sns.set(style="darkgrid", palette="muted", color_codes=True)

# Plot the boxsplots
ax = sns.boxplot(data=movies3, x='Genre', y='Gross % US', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Add in points to show each observation
sns.stripplot(x='Genre', y='Gross % US', data=movies3, jitter=True, size=6,linewidth=0,  hue = 'Studio', alpha=0.7)

ax.axes.set_title('"Domestic Gross % by Genre"',fontsize=30)
ax.set_xlabel('Genre',fontsize=20)
ax.set_ylabel('Gross % US',fontsize=20)

# Define where to place the legend
ax.legend(bbox_to_anchor=(1.05, 1),loc=2,  borderaxespad=0)
```
